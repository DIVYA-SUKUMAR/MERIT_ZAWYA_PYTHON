```python
import time
import selenium
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import pandas as pd
import os

from pytz import timezone 
from datetime import datetime
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
```

```python
driverPath=r"C:\Users\Merit\Desktop\chromedriver.exe"
driver = webdriver.Chrome(driverPath)
driver.get("https://www.zawya.com/mena/en/companies/list/cso/?keyword=Islamic%20Finance%20&country=&sector=&ownership=&resultsPerPage=10&page=2")
driver.maximize_window()

# CREATING EMPTY LIST 
Company_Name=[]
Country_of_Incorporation=[]
Estabilished_In=[]
Primary_Sector=[]
Primary_Address=[]
Region=[]
Email=[]
Website_Link=[]    
PageUrl=[]

# EXECUTION OF THE SCRIPT
for x in range(3,15):
    for i in range (1,11):
        factdata=[]  
        click = driver.find_element_by_xpath("//a[contains(text(),'LEARN MORE')]//following::a["+str(i)+"]")
        driver.execute_script("arguments[0].click();",click) 
        PageUrl.append(driver.current_url)
        time.sleep(2)

        name= driver.find_elements_by_xpath("//h1[@class='page-title']") 
        for j in name:
            Company_Name.append(j.text)
        factdata=driver.find_elements_by_xpath("//h3[contains(text(),'fast facts')]")
        if (len(factdata)==1):
            country=driver.find_elements_by_xpath("//h3[contains(text(),'fast facts') ]//following::strong[1]")
            for a in country:
                Country_of_Incorporation.append(a.text)
            estabilished=driver.find_elements_by_xpath("//h3[contains(text(),'fast facts')]//following::strong[2]")
            if(len(estabilished)==4):
                for c in estabilished:
                    Estabilished_In.append(c.text)
            else:
                Estabilished_In.append("Null")
            sector=driver.find_elements_by_xpath("//h3[contains(text(),'fast facts')]//following::strong[3]")
            for b in sector:
                Primary_Sector.append(b.text)
        else:
            country=driver.find_elements_by_xpath("//h2[contains(text(),'About')]//following::a[1]")
            for a in country:
                Country_of_Incorporation.append(a.text)
            estabilished=driver.find_elements_by_xpath("//h2[contains(text(),'About')]//following::div[@class='col-md-9'][2]") 
            if(len(estabilished)==4):
                for c in estabilished:
                    Estabilished_In.append(c.text)
            else:
                Estabilished_In.append("Null")
            sector=driver.find_elements_by_xpath("//h2[contains(text(),'About')]//following::a[2]")
            for b in sector:
                Primary_Sector.append(b.text)
        try:
            address=driver.find_elements_by_xpath("//h5//following::address")
            if(len(address)!=0):
                for l in address:
                    Primary_Address.append(l.text)
            else:
                Primary_Address.append("Null")
            region=driver.find_elements_by_xpath("//h5[contains(text(),'Region')]//following::strong[1]")
            if(len(region)!=0):
                for m in region:
                    Region.append(m.text)
            else:
                Region.append("Null")
            link=driver.find_elements_by_xpath(" //h5[contains(text(),'CONTACT')]//following::li[2]")
            if(len(link)!=0):
                for n in link:
                    Website_Link.append(n.text)
            else:
                Website_Link.append("Null")
        except:
            print("Stop")
        driver.back()     
    button = driver.find_element_by_xpath("//a[@class='page-link next fa fa-angle-right']")
    driver.execute_script("arguments[0].click();", button)
driver.close()
```

```python
# SPLITTING THE DATA
from pytz import timezone 
from datetime import datetime

zawya_12=pd.DataFrame()
zawya_12['SNo']=zawya_12.index
zawya_12.set_index('SNo',inplace=True)
zawya_12['Company_Name']=Company_Name
zawya_12['Country_of_Incorporation']=Country_of_Incorporation
zawya_12['Estabilished_In']=Estabilished_In
zawya_12['Primary_Sector']=Primary_Sector
zawya_12['Primary_Address']=Primary_Address
zawya_12['Region']=Region
zawya_12['Date_Time']= datetime.now(timezone("Asia/Kolkata")).strftime('%Y-%m-%d %H:%M:%S.%f')
zawya_12['Website_Link']=Website_Link
zawya_12['PageUrl']=PageUrl

zawya_12
```

```python
import os
import sqlite3
from flask import jsonify
from flask_api import FlaskAPI
from flask_restful import Api
from sqlalchemy import create_engine
from flask import request
import json
app=FlaskAPI(__name__)
api=Api(app)
c=create_engine('sqlite:///PROJECT_ZAWYA_Python1.db')

# ROUTE FOR CREATE

@app.route('/Zawya_create', methods=['GET'])
def create():
    try:
        con = c.connect()
        trans = con.begin()
        con.execute('CREATE TABLE IF NOT EXISTS Zawya_Tbl1(SNO INT PRIMARY KEY,Company_Name TEXT,Country_of_Incorporation TEXT,Estabilished_In TEXT,Primary_Sector TEXT,Primary_Address TEXT,Region TEXT,Website_Link TEXT,PageUrl TEXT,Timestamp DATETIME DEFAULT CURRENT_TIMESTAMP)')
        trans.commit()
        text = {"Message " : "Table is Created"}
        return text
    except Exception as error:
        text = {"Message " : "Table Creation Failed"}
        return text
    finally:
        con.close()

# ROUTE FOR INSERT

@app.route('/Zawya_insert', methods=['GET'])
def insert():
    con = c.connect()
    zawya_12.to_sql('Zawya_Tbl1', con,if_exists='replace')
    text = {"Message " : "Table is Inserted"}
    return text

# ROUTE FOR SELECT

@app.route('/Zawya_select', methods=['GET'])
def select():
    con = c.connect()
    sel = con.execute("select * from Zawya_Tbl1")
    return json.dumps([dict(r) for r in sel])

app.run()
```
